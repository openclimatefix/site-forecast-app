## Before running the app

1. You will need to be authenticated in AWS. If you've never done this before, [install aws](https://docs.aws.amazon.com/cli/latest/userguide/getting-started-install.html), then [login](https://docs.aws.amazon.com/signin/latest/userguide/command-line-sign-in.html#command-line-sign-in-local-development). You will also need to generate an access key for your machine through the AWS console in your browser; once this is doen, call `aws configure` in the terminal and enter both public and secret keys when prompted.
2. Open a connection to the database in your terminal. Amend the `DB_URL` below to the correct port (same as in your `ssh`). DB name, username, and password can be retrieved from AWS secrets (you can retrieve the DB_URI and edit it to point to localhost)
3. Amend the environment variables below.
